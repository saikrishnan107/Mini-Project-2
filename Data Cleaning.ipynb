# Forest

Conversion of sheet name to excel

```python
# Getting sheet names
sheet_names = ex1.sheet_names
# Read Data of all sheets into dictionary
sheets_dict = {sheet: ex1.parse(sheet) for sheet in sheet_names}
```

```python
combine_df = pd.concat(
    [df.assign(sheet=sheet_name) for sheet_name, df in sheets_dict.items()],
    ignore_index=True)
df = combine_df.drop(columns=['sheet'])
```

```python
df
```

# Handling Missing Data

```python
# Remove missing and unecessary columns
# Remove Sub Unit Code and AcceptedTSN
df.drop(columns=["Sub_Unit_Code"], inplace=True)  # Drop columns with all NaN values
df.dropna(subset=["AcceptedTSN"], inplace=True)  # Drop columns with all NaN values

# Imputation
df['Distance'].fillna(df['Distance'].mode()[0],inplace=True)
df['Sex'].fillna(df['Sex'].mode()[0],inplace=True)
df['ID_Method'].fillna(df['ID_Method'].mode()[0],inplace=True)
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```
